```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plotImage(image, cmap='gray'):
    try:
        plt.imshow(np.asarray(image), cmap=cmap, vmin=0, vmax=255)
    except:
        print("Error plotting the image", color='red')
```
