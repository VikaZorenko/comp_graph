```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pylab import *
%matplotlib inline
```

```python
def binarize(imageArr, threshold):
    return np.where(imageArr < threshold, 0, 255)
```

```python
from skimage import data
scanned = data.page()

plt.imshow(scanned, cmap='gray')
```

```python
binarized = binarize(scanned, 90)

plt.imshow(binarized, cmap='gray')
```

```python
scanned = array(Image.open('images/minions.png').convert('L'))

binarized = binarize(scanned, 92)

plt.imshow(binarized, cmap='gray')
```

```python
scanned = array(Image.open('images/MARBLES.bmp').convert('L'))

binarized = binarize(scanned, 80)

plt.imshow(binarized, cmap='gray')
```
