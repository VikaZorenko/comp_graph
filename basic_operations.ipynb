```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
```

```python
def plotImage(image, cmap='gray'):
    try:
        plt.imshow(np.asarray(image), cmap=cmap, vmin=0, vmax=255)
    except:
        print("Error plotting the image", color='red')
```

```python
img = Image.open('images/MARBLES.bmp')
plotImage(img)
```

```python
# Grayscale version
new_img = Image.open('images/MARBLES.bmp').convert('L')
plotImage(new_img)
```

```python
# Convert all jpg images to png
```

```python
import os
```

```python
file_list = ['images/minions.jpg']

for infile in file_list:
    outfile = os.path.splitext(infile)[0] + ".png"
    if infile != outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
            print("Cannot convert", infile)
```

```python
# Basic manipulations

img = Image.open('images/MARBLES.bmp')
plotImage(img)
```

```python
# Resize
smaller_img = img.resize((100, 100))
plotImage(smaller_img)
```

```python
# Crop
box = (100, 100, 400, 400)
region = img.crop(box)
plotImage(region)
```

```python
# Rotate
rotated_img = img.rotate(135)
plotImage(rotated_img)
```
